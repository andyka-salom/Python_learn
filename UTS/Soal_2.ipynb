```python
import numpy as np
import pandas as pd
```

```python
my_data = pd.read_csv('superstore.csv')
my_data.head(10)
```

```python
my_data['Order Date'] = pd.to_datetime(my_data["Order Date"])
```

```python
my_data['Year'] = my_data['Order Date'].dt.year
```

```python
data_2017 = my_data[my_data['Year'] == 2017]['Sales']
```

```python

#data_2017 = my_data[my_data['Order Date'].dt.year == 2017]['Sales']
```

**ALASAN PENGELOMPOKAN DATA MENGGUNAKAN KUARTIL**

Karena menurut saya pengelompokan menjadi 3 bagian akan terasa cepat dan mudah jika menggunakan kuartil, sebagaimana mestinya yang saya tau, bahwa Q1 dan Q3 adalah setengah bagian dari Median atau 25% 50% 75%.

```python
Q1 = data_2017.quantile(0.25)
Q2 = data_2017.median()
Q3 = data_2017.quantile(0.75)
```

```python
def segment_customer(row):
  if row['Sales'] > Q3:
    return 'High'
  elif row['Sales'] < Q1 :
    return 'Low'
  else :
    return 'Medium'
```

```python
my_data['segmentasi'] = my_data.apply(segment_customer, axis=1)
```

```python
my_data.head(10)
```

```python
jum_data_segmen_2017 = my_data[my_data['Year'] == 2017]['segmentasi'].value_counts()
```
